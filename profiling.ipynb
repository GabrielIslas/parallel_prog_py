```python
import pi_approx
import pi_approx_mp
import time
```

```python
test_N = [10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000]
```

```python
normal_times = []
for N in test_N:
    start = time.time_ns()
    pi_approx.approximate_pi(N)
    end = time.time_ns()
    normal_times.append(end - start)
```

```python
normal_times
```

```python
mp_times = []
for N in test_N:
    start = time.time_ns()
    pi_approx_mp.main(N)
    end = time.time_ns()
    mp_times.append(end - start)
```

```python
mp_times
```

```python
mpi_times_seconds = [2.367, 2.295, 2.297, 2.370, 2.335, 2.294, 2.325, 3.477]
```

```python
# Converting to nanoseconds like everything else
mpi_times = [mpi_time * 1000000000 for mpi_time in mpi_times_seconds]
```

```python
mpi_times
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(test_N, normal_times, label='normal')  
plt.plot(test_N, mp_times, label='multiprocessing')  
plt.plot(test_N, mpi_times, label='mpi4py')
plt.title('Runtime comparison')  
plt.xlabel('N')                
plt.ylabel('Time (ns)')
plt.legend()
plt.show()
```
